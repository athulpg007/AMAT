```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
class Cannon:
    
    def __init__(self):
        pass
    
    def EOM(self, t, y, acc): 
        return [y[1], acc]
    
    def solve_trajectory(self, t_sec, dt, y0, acc, events):
        tbar = np.arange(0, t_sec + dt, dt)
        sol = solve_ivp(self.EOM, (0, tbar[-1]), y0,  args=(acc,), t_eval=tbar, events=events)
        
        self.t = sol.t
        self.y = sol.y[0, :]
        
    def reach_apex(self, t, y, acc):
        return y[1]
    
    def reach_100_m_below(self, t, y, acc):
        return y[0] + 100
    
    reach_apex.terminal = True
    reach_100_m_below.terminal = True
```

```python
cannon = Cannon()
event1 = cannon.reach_apex
event2 = cannon.reach_100_m_below
cannon.solve_trajectory(t_sec=100, dt=1, y0=[0, 10], acc=-0.5, events=event2)
```

```python
plt.plot(cannon.t, cannon.y)
plt.show()
```
