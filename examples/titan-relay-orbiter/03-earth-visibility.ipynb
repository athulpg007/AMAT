```python
from AMAT.visibility import LanderToPlanet, LanderToOrbiter, OrbiterToPlanet
```

```python
visibility1 = LanderToPlanet(
    observer_planet="TITAN", target_planet="EARTH", latitude=3.0, date="2034-12-31 00:00:00",
    ephem_file='../../spice-data/de432s.bsp'
)

visibility2 = LanderToPlanet(
    observer_planet="TITAN", target_planet="EARTH", latitude=80.0, date="2034-12-31 00:00:00",
    ephem_file='../../spice-data/de432s.bsp'
)

visibility3 = LanderToPlanet(
    observer_planet="TITAN", target_planet="EARTH", latitude=-80.0, date="2034-12-31 00:00:00",
    ephem_file='../../spice-data/de432s.bsp'
)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8.5, 3.5))
plt.rc('font',family='Times New Roman')
params = {'mathtext.default': 'regular' }
plt.rcParams.update(params)

plt.plot(visibility2.t_array/86400, visibility2.elevation_array, 'k--', linewidth=2.0, label = '80 deg. N')
plt.plot(visibility1.t_array/86400, visibility1.elevation_array, 'b-', linewidth=2.0, label = '3 deg. N')
plt.plot(visibility3.t_array/86400, visibility3.elevation_array, 'r-.', linewidth=2.0, label = '80 deg. S')

plt.xlabel('Time, days.',fontsize=16)
plt.ylabel('Earth elevation, deg.',fontsize=16)
ax=plt.gca()
ax.tick_params(direction='in')
ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.tick_params(axis='x',labelsize=16)
ax.tick_params(axis='y',labelsize=16)
plt.grid(linestyle='dotted', linewidth=0.5)

plt.legend(loc='upper right', fontsize=12)

plt.savefig('../../data/joaaa-titan-orbiter/dte-elevation.png', dpi= 300,bbox_inches='tight')
plt.savefig('../../data/joaaa-titan-orbiter/dte-elevation.pdf', dpi=300,bbox_inches='tight')
plt.savefig('../../data/joaaa-titan-orbiter/dte-elevation.eps', dpi=300,bbox_inches='tight')

plt.show()
```
