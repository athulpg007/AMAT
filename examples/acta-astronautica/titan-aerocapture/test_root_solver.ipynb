```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
```

```python
def compute_E(e, M):
    root = optimize.newton(lambda E: E - e*np.sin(E) - M, x0=M, rtol=1e-6)
    return root
```

```python
compute_E(0.25, np.pi)
```

```python
M_array = np.linspace(0, 2*np.pi, 101)

E1_array = np.array([])
E2_array = np.array([])
E3_array = np.array([])

for M in M_array:
    E1 = compute_E(0.00, M)
    E2 = compute_E(0.50, M)
    E3 = compute_E(0.99, M)
    E1_array = np.append(E1_array, E1)
    E2_array = np.append(E2_array, E2)
    E3_array = np.append(E3_array, E3)
```

```python
plt.figure(figsize=(8,8))
plt.plot(E1_array, M_array, 'r--')
plt.plot(E2_array, M_array, 'g--')
plt.plot(E3_array, M_array, 'b--')
plt.show()
```

```python
def compute_true_anomaly(e, E):
    return 2*np.arctan(np.sqrt((1+e)/(1-e))*np.tan(E/2))
```

```python
theta_star = compute_true_anomaly(0.99, E3_array)
```
